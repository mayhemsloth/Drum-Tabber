### Audio Data Pre-processing and Explanation
There are many models used in ADT but **I am most interested in using convolutional neural networks (CNNs) and perhaps convolutional recurrent neural networks (CRNNs)** as they seem to perform the best in the most recent literature. For CNNs and CRNNs almost always the first step in preprocessing the audio so that features can be properly extracted by the NN is **creating a log mel-spectrogram**. This [blog post here](https://towardsdatascience.com/getting-to-know-the-mel-spectrogram-31bca3e2d9d0) explains well what a log mel-spectrogram is, but I will summarize it now, from the beginning of sound, so that I can internally understand it better from its base parts. 

Sound is simply a longitudinal pressure wave, transmitted through vibrations in a medium (usually air). That pressure wave can be measured and characterized as an amplitude as a function of time. Because computer electronics are based around digital objects, a continuous wave (like sound) must be **sampled** to represent it as a discrete sequence of amplitude numbers and store it as a digital object. The *rate* at which a continuous wave is sampled will affect the eventual apparent output shape of that wave. So what *rate* is chosen to sample a wave to make it discrete? Because humans have evolved to hear sound in air for the frequencies between 20 Hz and ~20,000 Hz, we can assume that we don't care about any frequencies above 20,000 Hz. This is our [Nyquist frequency](https://en.wikipedia.org/wiki/Nyquist_frequency), so we know we need to sample at >40,000 Hz in order to safely capture all the *frequency* information needed in a discretized sound wave to faithfully recreate it back to a continuous sound wave without humans detecting a difference. The standard audio sampling rate used pretty much everywhere is indeed **44,100 Hz**. Furthermore, in order to properly capture the dynamics of a small and large amplitudes of these pressure waves, the *bit depth* is also often defined (that is, how many different integer numbers you are allowed to use to map the continuous amplitude). The bit depth is pretty much standard at 16 bits (CD quality) or 24 bits (DVD/BluRay audio). 

Digital audio is thus an ordered 1D array of integers, that can, with the properly sampling rate, be converted back to a "continuous" wave and sound through speakers. However, **an ordered 1D array of integers is usually *not* the thing that is fed into a NN directly.** [Note that this case is different from computer vision, where the digital representation of images are often directly fed into NNs (rather than their k-space interpretation).] Any continuous wave object displayed in the time-domain can be transformed using **Fourier series transformations** to have it be represented in the frequency domain, which displays the relative prevelance of the different frequencies within that wave object. The same could be said for discrete wave objects and an appropriately changed Fourier transform function. For ADT and other music information retrieval tasks, the Fourier transform of an *entire song* at once is not that useful, because it doesn't describe the drum events or notes played as a function of time (just as a function of the entire song). If the goal was to know which frequencies occurred in some small time frame, you could slice the song up into tiny frames and *then* do a Fourier transform on *each* of these frames. Indeed, that is exactly what occurs with a [short-time Fourier transform](https://en.wikipedia.org/wiki/Short-time_Fourier_transform) (STFT). Remember that we are working under the assumption that drum events are "instantaneous" events. Slicing up a song into tiny durations of time, and *then* taking the Fourier transforms of those tiny slices is a reasonable approach to attempting to find "instantaneous" events. In reality, drum events do have some characteristic "duration", so ensuring that your transform slices aren't too *small* for these durations is an important consideration.  

So you now have a representation of the audio, as a function of discrete time steps, for which frequencies occur at (roughly) that specific time. **This representation is called a [spectrogram](https://en.wikipedia.org/wiki/Spectrogram)**, defined as a visual representation of a spectrum of frequencies of a signal as the signal varies with time. On the x-axis is time, the y-axis is the frequencies (in Hz), and then each point on that graph/matrix has a value, and is colored a certain color depending on the prevelance of that frequency, to produce a heat map. A **log spectrogram** is simply representing the y-axis as a log axis (to more accurately represent how humans hear) *and* the color axis on a log axis (to more accurately represent decibels, dB, which is also how humans hear). So what is a **log *mel*-spectrogram?** Well, to *even further* accurately represent how humans hear sound, we can transform the frequency axis (in Hz) into a different unit called mels (short for melody?). The [mel scale](https://en.wikipedia.org/wiki/Mel_scale) is "a perceptual scale of pitches judged by listeners to be equal in distance from one another". This mel scale is effectively a non-linear frequency axis transformation that takes the Hz frequencies and puts them into "equally spaced" bins that are "equal in distance" according to how humans hear them. Once the y-axis has been transformed from Hz to mels, then you are left with a log mel-spectrogram of some audio clip. As a data structure, a mel-spectrogram is a 2D array, where the dimensions are determined by the length of the audio clip (in seconds), the hop length (number of samples between successive frames), the window length (how many samples for each STFT), and the number of mel bins.

#### Making log mel-spectrograms from MAT dataframes and target labeling considerations
The point of the music-aligned tab dataframes is that I can easily provide a drum piece label to a slice of music, in the hopes that the music/label pairing could potentially be used as an example in a training, development, and/or test set. Well, let's think more deeply about this potential setup, and compare it to previous papers or data sets that utilize spectrograms and CNNs or CRNNs for model building. 

Let's say we already have a model that "works" in the ideal case: that the model takes in a song (and maybe its BPM or other simple information about the song) and outputs a drum tab (either with the measure character or not). In order to properly place the onset of a drum event, the model would need to check sufficiently small slices of the song, just as the training data had fed the model sufficiently small slices of the the training songs. **However there's a potential problem here and it has to do with the "phase" associated with tabbing a song and slicing it the way that I have done.** Currently, given some single 16th note grid event, the song slice associated with that 16th note event starts at the "start" of that 16th note (as determined by the human assigning the first drum note onset) and ends at the beginning of the next 16th note event. This establishes some "phase" of the tab. In fact, an equally valid way to slice the song and assign the drum event labels could be *some amount* of samples starting forward in time, instead of at the "start" of the 16th note grid event, up to half the length of the 16th note resolution. A better way to think about this is to think about what happens when a song gets put into an ideal model. If you slice the song at the wrong "phase" you may never encounter *any* drum event, as determined by the model, because the SFTFs that make up the spectrogram will consistently look *different* to the model than those slices that the model was trained on. The solution to this problem is to make the hop size so small that no matter where the slicing of the song *starts* (to be transformed into STFTs and spectrograms) that a certain slice *will* be captured to appropriately be interpreted by the model as having a drum note onset in it. 

Indeed, this is the case that previous papers describe. For example, in "Drum Transcription via Joint Beat and Drum Modeling Using Convolutional Recurrent Neural Networks" by Vogl et. al., a paper which uses log spectrograms to train RNNs, CNNs, and CRNNs, the window size in the STFTs was 2048 samples, or **46.4 ms** long, and they did this STFT every **10 ms** which is the frame size, and then assigned a drum note onset to one single frame. Checking a song every 10 ms will certainly find a window that entirely contains any potential drum onset in the song. In comparison, for a 150 BPM song, the 16th note grid resolution is 100 ms. Checking every 10 ms, in my opinion, is potentially overkill, but I guess during training and testing I can see what changing that hyperparameter does to the performance of the model. 

So it seems like for training NNs using spectrograms, the target functions are generated by setting frames of a signal with the same frame rate as the input features to the correct class label. Basically you are assigning each hop size (each frame) to a specific label. This practice makes perfect sense to me. The main decision that you need to make is **how many frames do you assign as being "correct"**, and **how long are your window sizes**, and do those two decisions interact with each other? 

I think that I want to do the following: I want to have a song go into the model accompanied by its BPM. (Alternatively, you could use a librosa built-in function to approximate the BPM becasue the exactness doesn't really matter). In this way you can make the STFT window that makes up the spectrogram dependent on the BPM, while the total frame number will *not* be a function of the BPM. For example, if you give it two songs that are each 4 minutes, but one has BPM of 120 and another has BPM of 160, the total number of frames that it will process will be the same because that is only dependent on the hop size and the total duration. However the STFT will yield be different because the window sizes will be different. The point of this is to reflect how the labels are produced: the information of the labels is gathered only based on a BPM basis. That is, the window sizes of the training data itself will be different.

Let's say a song's BPM is 150, so the 16th note resolution would be 15000/150 = 100 ms, which is my current plan for the calculation of the STFT window size. Let's say we have a slice that has some drum event in it. If a STFT has a length of 100 ms, and then we have a hop size of 20 ms (882 samples), that means that *some* part of any one 16th note slice of the song would be contained in (100 ms / 20 ms)x2 = ~10 STFTs. For example, if our 16th note grid slice starts at 350 ms and ends at 450 ms, and our hops are 20 ms along the 20 ms intervals, then the *first* time that a STFT would contain any part of the 16th grid drum event would be at the 260 ms-360 ms STFT slice, with a true overlap ratio of 10% = (360-350)/100. After that, some part of the 16th note drum event would be found in the STFT slices starting at 280 (30%), 300 (50%), 320 (70%), 340 (90%), 360 (90%), 380 (70%), 400 (50%), 420 (30%), and 440 (10%). If this song is being used as a training sample, which of these STFT slices *should* be assigned to a drum event label, and which should not be? The drum events are most likely to be at the beginning of the 16th note duration, although in reality the entirety of the 16th note duration is "useful" to identify the drum event. I probaly will decide that any STFT slice that overlaps some percentage (for example, at least 80%) of the drum note event slice would count as a positive in the label, but this is probably something that I can tinker around with later on. Alternatively, since I am assuming that I am slicing the song at the *beginning* of the drum notes, I could include the one that is closest to the starting time, and a a couple hops before that as well. This method would ensure that for any positive label in the STFT slices, *at least* the beginning of a 16th note slice would be included in that STFT slice.

### Constructing the X and Y Training Set
From here on, the X and Y terms will refer to the entire X matrix training data (that is directly fed into the future NN architecture) and the entire Y matrix classification and labels data (the entire set of "correct" labels that is used in future NN architecture). Note that X and Y will have the same number of examples: that is, every X example has a Y label associated with it. 

The idea is to construct the X and Y training set from the current FullSet dataframe. **Instead of directly converting the slices in the FullSet to a spectrogram, a function will be constructed  to manipulate the tab slices to make more windows for the spectrogram (right now, the plan is for 4 windows per tab slice), but only one window will be labeled as a positive drum event.** Note this may change in the future based on the performance of the model. Perhaps the last "quarter" spectro slice will be labeled properly by the following tab slice, because it most likely contains the drum event of that tab slice. 

```python
def create_XY(FullSet_df, sr= 44100, n_mels = 150, include_LR = False, n_spectro_slices = 4, include_differential = False):
    """
    High level function that takes in a FullSet dataframe (multi-index, still ordered) and uses the 'song slice' label to make X,
    and uses the tab labels to make Y.
    
    Args:
        FullSet_df [Dataframe]: has 'song slice' column with the raw audio, and then the rest are columns containing the drum event labels
        sr [int]: Default = 44100. The sampling rate of all the songs in the FullSet_df. Probably should be known at the song level but whatever. 
        n_mels [int]: Default = 150. This is the number of mel bins when making the mel spectrograms later on 
        include_LR [bool]: Default = False. The FullSet_df contains the stereo (2 channels) version of the songs. Building spectrograms using librosa can only
                          use mono audio samples. When False, each song will be converted to mono and only that will be processed into X
                          When True, each song will be converted to mono still, AND the Left and Right channels individually will ALSO be processed 
                          to produce effectively 3x as many examples in the data. 
        n_spectro_slices [int]: Default = 4. The number of spectrogram slices that is created per tab slice
        include_differential [bool]: Default = False. Decides whether to include the first differential over time of the Spectrogram plot 
    
    Returns:
        np.array: X, the input data set
        np.array: Y, the correct label data set
        dict: note_dict, a description from the encoder that describes a bunch of information about the X and Y dataset (what columns mean, dimensions, etc)
    """
    
    note_dict = {'n_mels' : n_mels, 'sr': sr, 'n_spectro_slices' : n_spectro_slices, 'include_LR': include_LR, 'include_differential' : include_differential}  # used for information aggregation that may be lost to the user when the dataframe structure is no longer used
    XY_dict = {}    # keys should be song name, values should be the XY dataframe for that song, to construct a multi-index XY DF later
    
    # encodes the drum labels in the columns of the FullSet into "one hot" (multi-label) columns based on the columns and their labels
    FullSet_encoded = Y_encoder(FullSet_df)
    
    # go through each song in FullSet_df and construct a NEW dataframe of the spectrograms (X) and their proper labels (Y) 
    for song in list(FullSet_encoded.index.levels[0]):    # loop through each song in the multi-index FullSet_df
        cs_df = FullSet_encoded.loc[(song,),:]            # grabs the current song's df rows only
        list_of_row_dicts = []                     # construct the dataframe through a list of dictionaries that describe each row
        tab_labels_dict = cs_df.drop(columns= ['song slice']).to_dict('index')   # a dictionary containing the index values as keys and as values is another dictionary, containing the column names as keys and their corresponding values (that used to be inside the df for that index) as values
        for index in range(len(cs_df.index)-1):    # doesn't do the last index of cs_df (due to the decision that was made for how to process the STFTs)
            concat_with_next = np.concatenate((cs_df['song slice'][index], cs_df['song slice'][index+1]), axis = 0)  # concat next song slice with current index to produce spectro slices smoothly into the next index
            init_n = cs_df['song slice'][index].shape[0]
            next_n = cs_df['song slice'][index+1].shape[0]
            L_concat = concat_with_next[:,0]  # grabs a 1D array of the Left channel of the concatenated song slices
            R_concat = concat_with_next[:,1]  # grabs a 1D array of the Right channel of the concatenated song slices
            mono_concat = lb.core.to_mono(concat_with_next.T)  # creats a mono audio channel of the concatenated song slices
            for spectro_num in range(n_spectro_slices):  # 0, 1, 2, 3 when using default
                list_of_row_dicts.append(spectro_row(spectro_num, n_spectro_slices, init_n, next_n, mono_concat, L_concat, R_concat, include_LR, tab_labels_dict, index, sr, n_mels))  # create the spectro row dictionary entry and append it to the running list
        print("About to update XY_dict with", song)
        # update the XY_dict with a song title as key, and then a DataFrame of that song's spectro slices as the Value
        XY_dict.update({song : pd.DataFrame(list_of_row_dicts)})
        
    # create the MultiIndex dataframe XY_df and set the NaNs to 0
    XY_df = pd.concat(XY_dict, axis=0, ignore_index = False, join = 'outer', sort = False)    # multiIndexed by song (by nature of its construction)
    XY_df = XY_df.fillna(0)
    
    """ EXTRACT X and Y """
    # extract the X matrix and song information about it
    X_songs = list(XY_df.index.levels[0])     # list of songs
    X_song_order_dict = {idx : val for idx, val in enumerate(X_songs)}       # dict for order of those songs
    X_song_length_dict = {song: len(XY_df.loc[(song,),:].index) for song in X_songs}   # dict for length of songs
    
    index_range_dict = {}    # dict that contains the song's index ranges within the final X matrix, construction below
    counter = 0
    for song in X_songs:
        index_range = [counter, counter + X_song_length_dict[song]-1]
        index_range_dict.update({song:index_range})           
        counter = counter + X_song_length_dict[song]
    note_dict.update({'X song order' : X_song_order_dict, 'X song length' : X_song_length_dict, 'X song index range': index_range_dict})  # update the main note_dict to include these new dictionaries
    
    X_names = [col for col in list(XY_df.columns) if 'spectro' in col] # gets a list of spectro columns
    X_column_order_dict = {idx : val for idx, val in enumerate(X_names)}
    note_dict.update({'X spectro index order' : X_column_order_dict})
    
    X = X_creator(XY_df, include_differential)
    
    # extract the Y matrix and information about it
    Y_names = [col for col in list(XY_df.columns) if 'spectro' not in col]  # gets a list of non-spectro columns, in the same order as in the dataframe from which they come
    Y_column_order_dict = {idx: val for idx, val in enumerate(Y_names)}  # creates a dictionary of the index values of the Y columns as keys and the Y column names as values
    note_dict.update({'Y column index order' : Y_column_order_dict})
    Y = XY_df.drop(columns =  [col for col in list(XY_df.columns) if 'spectro' in col]).astype(int).to_numpy() # drops the columns that have "spectro" in the name, casts all inside values (0s & 1s) as ints, then outputs to a numpy array
    
    return X, Y, note_dict
```

#### X_creator function
The Y matrix will need its own function for encoding, but the actual creation after that is a just a one-liner. The X matrix is slightly more complicated because of the choice given to the user to include the first time differential of the log melspectrogram into the X function, so it will get its own function to output the X matrix. 

```python
def X_creator(XY_df, include_differential):
    """
    asdf
    
    Args:
        XY_df [Dataframe]: the XY_df that is created inside create_XY function as a processing step. Contains the spectro slices and the encoded Y columns
        include_differential [bool]: boolean that determines if the first time derivative of the spetrogram matrix will be included in the X data set
        
    Returns:
        np.array: the X data set in a np.array of some dimensionality (depending mainly on n_mels [already encoded in length of spectros], 
                  include_differential, and the number of original examples of the 16th note drum tabs)
    """
    
    # Note we need to do this in a for loop instead of accessing all the data at once because the np.max part of the power_to_db takes into account the values of the spectrograms: 
    # We do NOT want to scale to ALL songs at the same time, because they are originally mixed/mastered in different ways and have different max volumes in the song
    spectro_cols = [col for col in list(XY_df.columns) if 'spectro' in col] # gets a list of spectro column names
    song_spectros_list = []                # build up the song_spectros to then concatenate them all together after the for loop
    for song in list(XY_df.index.levels[0]):
        spectros_from_columns_list = []
        for column in spectro_cols:      # loops through all the spectro columns in order 
            spectro_list = XY_df.loc[(song,), column].to_list()  # grabs a list of the current song's current spectrogram column
            full_song_spectro = np.concatenate(spectro_list, axis=1)  # concat all the numpy arrays in the list, along the correct axis
            # full_song_spectro.shape =  (n_mels, n_spectro_rows) where n_spectro_rows = n_spectro_slices*(len(tab_df)-1) because last index is thrown out
            full_song_db = lb.power_to_db(full_song_spectro, ref = np.max) # use entire song to convert to log mel spectrogram 
            if include_differential:
                first_time_derivative = lb.feature.delta(full_song_db, order=1)    # calculate the first time derivative of the full song spectrogram
                # first_time_derivative.shape = (n_mels, n_spectro_rows) SAME AS full_song_spectro
                full_song_db_and_ftd = np.concatenate([full_song_db, first_time_derivative], axis=0)
                # full_song_db_and_ftd.shape = (2* n_mels, n_spectro_row) with first_time derivative being attached to END of the normal log mel spectrogram (n_mels of spectro, then n_mels of ftd)
                spectros_from_columns_list.append(full_song_db_and_ftd)
            else:
                spectros_from_columns_list.append(full_song_db)
        song_spectros = np.stack(spectros_from_columns_list, axis = -1) # will produce either a (n_total, n_spectro_rows, 1) np.array for mono only, or (n_total, n_spectro_rows, 3) np.array for include_LR = True
        song_spectros_list.append(song_spectros)
    
    X = np.concatenate(song_spectros_list, axis = 1) # concatenate the list of full song spectros into one numpy array X
    
    return X
```

#### Y_encoder function
The labels are still in a roughly tab-readable format, with characters representing a drum event. For the Y label matrix, we need to convert that to a one hot representation, where different columns represent the different possible classes and the values of 0 (no event present) and 1 (event present) for each row. We have essentially already done most of the work for an encoder with the cleaning and collapsing of classes, so this function turned out to be pretty simple to code. Note that we are working in a **multi-label** system, so any event can be **multiple** drum events (because the bass, drum and cymbals can all be hit at the same time). So the "one hot" representation isn't entirely accurate, but it's one hot-like. 

```python
def Y_encoder(FullSet_df):
    """
    Encoder for encoding the Y labels of the FullSet dataframe. Encoded as a multi-label one hot vector: that is, a one hot vector that can be 1 in multiple classes for each example
    Note that the labels from the time-keeping line are classifed as "beats" and "downbeats", not "offbeats" and "downbeats"
    Note that the labels names are "encoded" into the column names of the dataframe, regardless of how many labels you might have in each column originally  
    
    Args:
        FullSet_df [Dataframe]: the FullSet_df that has been passed through the cleaning and collapsing classes functions already
        
    Returns:
        Dataframe: an encoded version of the FullSet_df
    """
    tk_label, measure_char, blank_char = get_special_chars()  # get blank_char mainly
    
    col_list = list(FullSet_df.drop(columns = ['song slice']).columns)
    for col in col_list:                     # goes through all the column names except 'song slice'
        uniques = [uniq for uniq in FullSet_df[col].unique() if uniq is not blank_char] # list of unique values found in current column
        if col == 'tk':                   # we are in the time-keeping column case, so we'll hard code handle this
            FullSet_df['tk_beat'] = FullSet_df['tk'].apply(lambda row1: 1 if row1 != blank_char else 0)   # create a new tk_beat column if there is any non blank char
            FullSet_df['tk_downbeat'] = FullSet_df['tk'].apply(lambda row2: 1 if row2 == 'C' else 0)    # create a new tk_downbeat column where there is capital C, the previously HARD CODED downbeat label     
        else:     # we are in all the other column cases, so we have to generically code this
            for label in uniques:   # we treat all the different labels as separate, regardless of how many we have on each line
                FullSet_df[col + '_' + label] = FullSet_df[col].apply(lambda row3: 1 if row3 == label else 0)  # create a column named like "BD_o" where it is 1 any time you have that label and 0 elsewhere
        FullSet_df = FullSet_df.drop(columns = [col])    # drop the column because we no longer need it as it has been encoded properly
                
    return FullSet_df
    
```

#### spectro_row function
Function specifically made to offload code into a different function. This function creates one single spectrogram row and puts it into a dictionary, which is then built up into a list to create a lsit of dicts which make up the dataframe later on. 

```python
def spectro_row(spectro_num, n_spectro_slices, init_n, next_n, mono_concat, L_concat, R_concat, include_LR, tab_labels_dict, index, sr, n_mels):
    """
    Lower level function used inside create_XY to create a spectrogram row dictionary to be appended to a list and built up into a Dataframe later
    
    Args:
        It's not worth typing these out. I wanted to compartmentalize some of my code in the high level function so it needed access to a bunch of vars
        
    Returns:
        dict: the current spectrogram row dictionary entry
    """
    
    spectro_row_dict = {}    # the row dictionary to update / build up 
    spectro_frac = spectro_num/n_spectro_slices  # which fraction the spectro number is at w.r.t. the total spectro slices
    
    # do mono concat first because all cases will use the mono spectro data
    spectro_slice = mono_concat[int((spectro_frac) * init_n) : init_n + int((spectro_frac)*next_n)]
    mono_S = lb.feature.melspectrogram(spectro_slice, sr=sr, n_fft = len(spectro_slice), center = False, n_mels = n_mels)
    spectro_row_dict.update({"spectro_mono": mono_S})  # "spectro_mono" will be the column name that holds the spectrograms created from the mono channel
    
    if include_LR: # if True, we grab proper spectro slice for L and R concats and create spectrograms for them, then update the dictionary
        L_slice = L_concat[int((spectro_frac) * init_n) : init_n + int((spectro_frac)*next_n)]
        L_S = lb.feature.melspectrogram(np.asfortranarray(L_slice), sr=sr, n_fft = len(L_slice), center = False, n_mels = n_mels)
        R_slice = R_concat[int((spectro_frac) * init_n) : init_n + int((spectro_frac)*next_n)]
        R_S = lb.feature.melspectrogram(np.asfortranarray(R_slice), sr=sr, n_fft = len(R_slice), center = False, n_mels = n_mels)
        spectro_row_dict.update({"spectro_left": L_S, "spectro_right" : R_S})  # note the column names of "spectro_left" and "spectro_right"
    if spectro_num == 0:      # in the case where the drum labels are given to the current spectro row
        spectro_row_dict.update(tab_labels_dict[index])
     
     # No other code is needed. If only a few columns are supplied in the creation of a df from a row dict, (compared to other rows)
     # it will still create a new row but all other columns will be NaN, which can simply be replaced with 0 after the df is constructed
     # because, currently, whenever the spectro_num is not 0 then all the other spectro slices get labels of 0s on the

    return spectro_row_dict
```
